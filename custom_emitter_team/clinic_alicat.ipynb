```python
import matplotlib.pyplot as plt

def plot_flow_controller_ranges(flow_controller_ranges, flow_controller_names):
    # Sort the flow controllers based on their lowest flow rates
    sorted_ranges = sorted(zip(flow_controller_ranges, flow_controller_names), key=lambda x: x[0][0])

    # Set up the figure and axis
    fig, ax = plt.subplots(figsize=(15, 8))
    ax.set_xscale('log')
    # Set the y-axis ticks and labels
    y_ticks = [i for i in range(len(sorted_ranges))]
    y_labels = [f'{name}' for (_, name) in sorted_ranges]
    ax.set_yticks(y_ticks)
    ax.set_yticklabels(y_labels)

    # Plot the horizontal lines for each flow controller range
    for idx, ((low_flow, high_flow), _) in enumerate(sorted_ranges):
        if flow_controller_names[idx] == '0.1' or flow_controller_names[idx] == '20' or flow_controller_names[idx] == '250' or flow_controller_names[idx] == '2000':
            ax.plot([low_flow, high_flow], [idx, idx], marker='o', markersize=8, label=f'{flow_controller_names[idx]}', color='blue')
        elif flow_controller_names[idx] == 'GOAL':
            ax.plot([low_flow, high_flow], [idx, idx], marker='o', markersize=8, label=f'{flow_controller_names[idx]}', color='green')
        else:
            ax.plot([low_flow, high_flow], [idx, idx], marker='o', markersize=8, label=f'{flow_controller_names[idx]}', color='grey')

        ax.text(low_flow, idx, f'{low_flow:.3f}', ha='right', va='center', fontsize=8)
        ax.text(high_flow, idx, f'{high_flow:.3f}', ha='left', va='center', fontsize=8)

    # Draw vertical lines at x = 0.034 and x = 283
    ax.vlines(x=[0.034, 283], ymin=0, ymax=len(sorted_ranges)-1, linestyle='dashed', color='red', label='Vertical Lines')

    # Add gridlines
    ax.grid(True, linestyle='--', alpha=0.7)

    # Set labels and title
    ax.set_xlabel('Flow Rate (SLM)')
    ax.set_title('Flow Controller Operating Ranges to Achieve 5% Error at Lowest Flow Rate')

    # Display legend
    ax.legend()

    # Show the plot
    plt.show()

# Example usage:
flow_controller_ranges = [(2, 8), (1, 5), (3, 10), (0.5, 6)]
custom_flow_controller_names = ['FC1', 'FC2', 'FC3', 'FC4']

flow_controller_ranges5per = [(0.05, 0.001), (0.1, 0.002), (0.2, 0.004), (0.5, 0.01), (1, 0.02), (2, 0.04), (5, 0.1),
                              (10, 0.2), (20, 0.4), (50, 2.38), (100, 4.76), (250, 11.90), (500, 23.81), (2000, 95.24),
                              (283, 0.034)]
custom_flow_controller_names2 = ['0.05', '0.1', '0.2', '0.5', '1', '2', '5', '10', '20', '50', '100', '250', '500', '2000', 'GOAL']

plot_flow_controller_ranges(flow_controller_ranges5per, custom_flow_controller_names2)
```
